# Create a dataset that we can analyse, which includes the results of the BTM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load tdm
tdm = pd.read_csv("/Users/jamiesanders/Desktop/PrintHistProj/output/results/BTM19_NONUMBERS_DEDUPED_AdStpwds__TOPIC_DOC_MAT.csv")

tdm.head()
```

```python
# Load processed data
processed = pd.read_csv("/Users/jamiesanders/Desktop/PrintHistProj/data/processed/estc_btm_prepped.csv")

processed.head()
```

```python
# Must equal TRUE!
processed.shape[0] == tdm.shape[1]
```

```python
# Returns most prev topic for each doc
most_prev_topic = tdm.idxmax(axis = 0).to_list()

# Add to dataframe three times
processed['top_topic_no'] = most_prev_topic

processed.head(3)
```

```python
# load original data
raw = pd.read_csv("/Users/jamiesanders/Desktop/PrintHistProj/data/raw/estc_with_pages.csv",
                        encoding = "ISO-8859-1")

raw = raw.set_index('system_number')

raw.head(3)
```

```python
# Join
processed = processed[['system_number','top_topic_no']]

final_results = raw.merge(processed.set_index('system_number'), how = 'inner', left_index = True, right_index = True)

final_results.head(20)
```

```python
final_results.to_csv("/Users/jamiesanders/Desktop/PrintHistProj/output/results/BTM19_RESULTS_NoLabel.csv")
```
